```python
# Import numpy and scipy
from pylab import *
%matplotlib inline 
from scipy.integrate import odeint
```

```python
# definitions
w = 1.0 # Well, we need to pick something. Might as well normalize the frequency. :-)
delta = 0.2 # damping constant,  see what happens when it gets very big, very small.
```

```python
def vhull_ODE(s,t):
# derives function for SH).  Need to modify for the hull without viscous damping
    g0 = s[1]
    g1 = -w*s[0]
    return array([g0,g1])
```

```python
def vhull_ODE2(s,t):
# derives function for SHO. Need to modify for v-hull ODE, with viscous damping added; 
#the damping term should include the damping coefficient delta
    g0 = s[1]
    g1 = -w*s[0]
    return array([g0,g1])
```

```python
#Note that the function is currently set as a simple harmonic oscillator function, not the V-hull boat

# Now set up the solutions.
time = linspace(0,20,500) # 500 points on interval [0,20]

# And now plot some things!

# Here's a low-amplitude one, should be approximately SHO.
A = 0.1 # Fairly small amplitude
yi = array([A,0.0])
answer = odeint(vhull_ODE, yi, time)
y = answer[:,0] # position is just the first row
plot(time,y, 'b-', label="Actual solution")
plot(time,A*cos(w*time), 'r-', label="SHO approximation")

A = 0.5 # Not small amplitude
yi = array([A,0.0])
answer = odeint(vhull_ODE, yi, time)
y = answer[:,0] # position is just the first row
plot(time,y, 'b-')
plot(time,A*cos(w*time), 'r-')
title("V-Hull Boat Solution")
xlabel("time")
ylabel("Position")
legend(loc="lower right")
show()
```
